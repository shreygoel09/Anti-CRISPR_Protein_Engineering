### Train/test split of anti-CRISPR sequences

```python
# read in raw data
df = pd.read_csv(dir + "/anti_crispr_cleaned.csv").drop(columns = ["Unnamed: 0", "index", "ID"])
```

```python
df
```

```python
# function to filter raw data based on select values
def model_prep(df, filter_value, new_df_name):
    filtered_df = df[df['Anti-CRISPR?'] == filter_value]
    filtered_df = filtered_df.reset_index(drop=True)
    filtered_df.name = new_df_name

    return filtered_df
```

```python
# filter for anti-crisprs and crisprs
crisp_df = model_prep(df, "CRISPR", "crisp_df")
anti_df = model_prep(df, "Anti", "anti_df")
```

```python
# save as csv
crisp_df.to_csv(dir + "/crisp_df.csv")
anti_df.to_csv(dir + "/anti_df.csv")
```
